```python
import pandas as pd
```

```python
everything = pd.read_csv('https://popcon.debian.org/by_inst',
    engine='python',
    delim_whitespace=True,
    header=0,
    comment='#',
    skipinitialspace=True,
    skipfooter=2,
    names=['rank','package','installed_by','used_by','old','updated_recently','no-files','extra'],
    usecols=['package','installed_by','used_by','updated_recently']
)
```

```python
# 5 most installed packages
everything.nlargest(n=5, columns=['installed_by'])
```

```python
# 5 most used packages
everything.nlargest(n=5, columns=['used_by'])
```
